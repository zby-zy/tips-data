# Scatter Plots

```python
# importing packages
import pandas as pd
import matplotlib.pyplot as plt
  
# reading the database
data = pd.read_csv("tips.csv")
  
# Scatter plot with day against tip
plt.scatter(data['day'], data['tip'])
  
# Adding Title to the Plot
plt.title("Scatter Plot")
  
# Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Tip')
  
plt.show()
```

```python
# importing packages
import pandas as pd
import matplotlib.pyplot as plt
  
# reading the database
data = pd.read_csv("tips.csv")
  
# Scatter plot with day against tip
plt.scatter(data['sex'], data['tip'])
  
# Adding Title to the Plot
plt.title("Scatter Plot")
  
# Setting the X and Y labels
plt.xlabel('Sex')
plt.ylabel('Tip')
  
plt.show()
```

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# reading the database
data = pd.read_csv("tips.csv")

sns.scatterplot(x='day', y='tip', data=data)

# Adding Title to the Plot
plt.title("Scatter Plot")

# Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Tip')

plt.show()
```

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# reading the database
data = pd.read_csv("tips.csv")

data.plot(kind = "scatter",x="total_bill",y = "tip")

# Adding Title to the Plot
plt.title("Scatter Plot")

# Setting the X and Y labels
plt.xlabel('Total Bill')
plt.ylabel('Tip')

plt.show()
```

```python
# importing packages
import pandas as pd
import matplotlib.pyplot as plt


# reading the database
data = pd.read_csv("tips.csv")

# Scatter plot with day against tip
plt.scatter(data['day'], data['tip'], c=data['size'],
			s=data['total_bill'])

# Adding Title to the Plot
plt.title("Scatter Plot")

# Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Tip')

plt.colorbar()

plt.show()
```

```python
# importing packages
import pandas as pd
import matplotlib.pyplot as plt

# reading the database
data = pd.read_csv("tips.csv")

# Scatter plot with day against tip
plt.plot(data['tip'])
plt.plot(data['size'])

# Adding Title to the Plot
plt.title("Scatter Plot")

# Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Tip')

plt.show()
```

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# reading the database
data = pd.read_csv("tips.csv")

sns.scatterplot(x='day', y='tip', data=data,
			hue='sex')

# Adding Title to the Plot
plt.title("Scatter Plot")

# Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Tip')

plt.show()
```

# Histogram

```python
# importing packages
import pandas as pd
import matplotlib.pyplot as plt

# reading the database
data = pd.read_csv("tips.csv")

# hostogram of total_bills
plt.hist(data['total_bill'])

# Adding Title to the Plot
plt.title("Histogram")

# Setting the X label
plt.xlabel('Total Bill')

plt.show()
```

# Line Plot

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# reading the database
data = pd.read_csv("tips.csv")

# draw lineplot
sns.lineplot(x="sex", y="total_bill", data=data)

# Adding Title to the Plot
plt.title('Line Plot')

# Setting the X and Y labels
plt.xlabel('Sex')
plt.ylabel('Total Bill')

plt.show()
```

# Bar Plot

```python
# importing packages
import pandas as pd
import matplotlib.pyplot as plt


# reading the database
data = pd.read_csv("tips.csv")

# Bar chart with day against tip
plt.bar(data['day'], data['tip'])

# Adding Title to the Plot
plt.title("Bar Chart")

# Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Tip')

plt.show()
```

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# reading the database
data = pd.read_csv("tips.csv")

sns.barplot(x='day',y='tip', data=data,
			hue='sex')

# Adding Title to the Plot
plt.title("Bar Chart")

# Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Tip')

plt.show()
```

# Heat map

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

f,ax = plt.subplots(figsize=(6, 6))
sns.heatmap(data.corr(), annot=True, linewidths=.5, fmt= '.1f',ax=ax)

# Adding Title to the Plot
plt.title("Heat map")

plt.show()
```

# Box Plot

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# reading the database
data = pd.read_csv("tips.csv")

data["total_bill"].plot(kind="box")

# Adding Title to the Plot
plt.title("Box Plot")

plt.show()
```
